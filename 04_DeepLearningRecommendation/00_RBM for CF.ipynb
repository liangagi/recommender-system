# Restricted Boltzmann Machines for Collaborative Filtering

## load package

```python
import random
import numpy as np

from surprise import NormalPredictor
from surprise.model_selection import GridSearchCV

from utils.MovieLens import MovieLens
from utils.Evaluator import Evaluator
from utils.RBMAlgorithm import RBMAlgorithm
```

```python
def LoadMovieLensData():
    ml = MovieLens()
    print("Loading movie ratings...")
    data = ml.loadMovieLensLatestSmall()
    print("\nComputing movie popularity ranks so we can measure novelty later...")
    rankings = ml.getPopularityRanks()
    return (ml, data, rankings)
```

```python
np.random.seed(0)
random.seed(0)
```

## read data

```python
# Load up common data set for the recommender algorithms
(ml, evaluationData, rankings) = LoadMovieLensData()
```

## compare RBM with random recommendations

```python
# Construct an Evaluator to, you know, evaluate them
evaluator = Evaluator(evaluationData, rankings)

#RBM
RBM = RBMAlgorithm(epochs=20)
evaluator.AddAlgorithm(RBM, "RBM")

# Just make random recommendations
Random = NormalPredictor()
evaluator.AddAlgorithm(Random, "Random")
```

```python
# Fight!
evaluator.Evaluate(True)

evaluator.SampleTopNRecs(ml)
```

## hyper-parameter tuning for RBM

```python
# Load up common data set for the recommender algorithms
(ml, evaluationData, rankings) = LoadMovieLensData()
```

```python
print("Searching for best parameters...")
param_grid = {'hiddenDim': [20, 10], 'learningRate': [0.1, 0.01]}
gs = GridSearchCV(RBMAlgorithm, param_grid, measures=['rmse', 'mae'], cv=3)

gs.fit(evaluationData)

# best RMSE score
print("Best RMSE score attained: ", gs.best_score['rmse'])

# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

## performance with best parameters

```python
# Construct an Evaluator to, you know, evaluate them
evaluator = Evaluator(evaluationData, rankings)

params = gs.best_params['rmse']
RBMtuned = RBMAlgorithm(hiddenDim = params['hiddenDim'], learningRate = params['learningRate'])
evaluator.AddAlgorithm(RBMtuned, "RBM - Tuned")

RBMUntuned = RBMAlgorithm()
evaluator.AddAlgorithm(RBMUntuned, "RBM - Untuned")

# Just make random recommendations
Random = NormalPredictor()
evaluator.AddAlgorithm(Random, "Random")
```

```python
# Fight!
evaluator.Evaluate(False)

evaluator.SampleTopNRecs(ml)
```
