# Auto-encoder for recommendation

## load package

```python
import random
import numpy as np

from surprise import NormalPredictor

from utils.MovieLens import MovieLens
from utils.Evaluator import Evaluator
from utils.AutoRecAlgorithm import AutoRecAlgorithm
```

```python
def LoadMovieLensData():
    ml = MovieLens()
    print("Loading movie ratings...")
    data = ml.loadMovieLensLatestSmall()
    print("\nComputing movie popularity ranks so we can measure novelty later...")
    rankings = ml.getPopularityRanks()
    return (ml, data, rankings)
```

```python
np.random.seed(0)
random.seed(0)
```

## read data

```python
# Load up common data set for the recommender algorithms
(ml, evaluationData, rankings) = LoadMovieLensData()
```

## compare Auto-encoder with random recommendation

```python
# Construct an Evaluator to, you know, evaluate them
evaluator = Evaluator(evaluationData, rankings)

#Autoencoder
AutoRec = AutoRecAlgorithm()
evaluator.AddAlgorithm(AutoRec, "AutoRec")

# Just make random recommendations
Random = NormalPredictor()
evaluator.AddAlgorithm(Random, "Random")
```

```python
# Fight!
evaluator.Evaluate(True)

evaluator.SampleTopNRecs(ml)
```
