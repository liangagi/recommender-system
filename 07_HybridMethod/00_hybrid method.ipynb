# Hybrid method
combine RBM and content based KNN for example

## load package

```python
import random
import numpy as np

from utils.MovieLens import MovieLens
from utils.RBMAlgorithm2 import RBMAlgorithm
from utils.ContentKNNAlgorithm import ContentKNNAlgorithm
from utils.HybridAlgorithm import HybridAlgorithm
from utils.Evaluator import Evaluator
```

```python
def LoadMovieLensData():
    ml = MovieLens()
    print("Loading movie ratings...")
    data = ml.loadMovieLensLatestSmall()
    print("\nComputing movie popularity ranks so we can measure novelty later...")
    rankings = ml.getPopularityRanks()
    return (ml, data, rankings)
```

```python
np.random.seed(0)
random.seed(0)
```

## read data

```python
# Load up common data set for the recommender algorithms
(ml, evaluationData, rankings) = LoadMovieLensData()

# Construct an Evaluator to, you know, evaluate them
evaluator = Evaluator(evaluationData, rankings)
```

## Compare hybrid method with other methods

```python
#Simple RBM
SimpleRBM = RBMAlgorithm(epochs=40)
#Content
ContentKNN = ContentKNNAlgorithm()

#Combine them with equal weights
Hybrid = HybridAlgorithm([SimpleRBM, ContentKNN], [0.5, 0.5])

evaluator.AddAlgorithm(SimpleRBM, "RBM")
evaluator.AddAlgorithm(ContentKNN, "ContentKNN")
evaluator.AddAlgorithm(Hybrid, "Hybrid")
```

```python
# Fight!
evaluator.Evaluate(False)

evaluator.SampleTopNRecs(ml)
```
